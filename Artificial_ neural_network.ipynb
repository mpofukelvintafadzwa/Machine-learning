```python
# In this jupyter notebook i will build an artificial neural network. # In building an Artificial Neural Network the first
# thing i will do is download the neccecary python libraries. The dataset here is for a bank and we would like to make predictions as
#to whether an employee is likely to leave the bank or not.

# Importing the libraries
import numpy as np
# Because numpy arrays are the only allowed inputs in the neural network as opposed to dataframes.
import matplotlib.pyplot as plt
# For plotting and visualizing the results.
import pandas as pd
# Because we want to download the data csv as a pandas dataframe which makes data management easier.

# The next step is to import the entire dataset (.csv file) as a pandas dataframe using the pd.read_csv
dataset = pd.read_csv('Churn_Modelling.csv')
# We look at the independant variables we want to include in our matrix features. In this dataset we have established that
# the independant variables which will contribute to our analysis run from 3-12. So we take the variable range 3:13 because the numpy array
# does not include the upper ound of the range.
#This step is to convert the data in the dataframe into a numpy array,we use the .iloc column to select the right index.
#You want to take all the lines from begining to left of the column all the way up to the end so you use :,
X = dataset.iloc[:, 3:13].values
# Matrix of features X (independant variables) has been created
# independant variable vector
y = dataset.iloc[:, 13].values

print (X)
```

```python
#splitting data into training set and test set. We have categorical
#variable in our matrix of features and we thus have to encode them
#We thus have a section which encodes any categorical data which we 
#have in our matrix of features. The label encoder encodes text into numbers.
#We will encode the categorical data before we split our data because when
#you do the split your matrix of features. Our matrix of features X has 2 sets of categorical
#that is the country and gender 
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
#we start off by encoding the country which has index 1 in the matrix of features
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
#we we then go on to encode the gender which has index 2 in the matrix of features
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
#Since the variables do not have precedence over one another we want to introduce some kind of dummy varible
#for the case where we have 3 label for our labelled categorical data. we will remove one column to avoid the dummy variable trap.
#One hot encoder creates these dummy variable coulumns.
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
#We will update this matrix by taking all the lines of the matrix and all the columns except the first one.
X = X[:, 1:]

print (X)
```

```python
#Splitting the data into training set and test set data. We train on 80 percent of the data and test on 20 percent.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#feature scaling is going to be important especially since there are going to be a large number of parallel computations.
#we also need to ensure that no variable dominates another. This is our last step in our data preprocessing.
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Building the ANN. We use some libraries designed for  deep learning which include the keras library built on the tensorflow
#backend meaning you need to download tensorflow first.
import keras
#The sequential module initializes the nueral network. We import this module from the keras.model 
from keras.models import Sequential
#The dense module adds layers to our network.We import this module from the keras.layers
from keras.layers import Dense
```

```python
#initializing the ANN as a sequence of layers. We define an object of the sequancial class
classifier = Sequential()
```

```python
# We use the add method to add new layers to our nueral network.# Adding the input layer and the first hidden layer
# we will intialize the weights uniformly. We will choose the relu activation function. input dim =11
# because we have 11 independant variables. 
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))

# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer. Make sure to check the number of categories for your dependant variable.
#if you have more than 2 categories for your dependant variable you need to use the softmax activation function.
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the ANN. loss function which needs to be optimized to get optimal weights of the neural network.
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Part 3 - Making predictions and evaluating the model

# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
